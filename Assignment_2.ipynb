1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.


```python
list=[100,200,300,400,500] #created a list of 5 integers

list.append(600) #Adds one element at the end of the list
list.extend([700,800]) #Adds multiple elements at the end of the list
list.insert(3,350) # Inserts an element at specified position
list.remove(200) # Removes the first occurrence of the specific element
popped_element=list.pop() #removes and return the last element 
print("Popped element: ",popped_element)
index_pos=list.index(400) #returns the index of the first occurrence of the specific element
print("Index Of ",400,"is",index_pos)
count_ele=list.count(500) #return the no.of times the specific element is appeared
print("count of ",500,"is",count_ele)
list.sort() #sorts the list
print("sorted list:",list)
list.reverse() #reverses the list
print("Reversed list: ",list)
list.clear() #clears all the elements in the list
print("after clearing list:",list)
```

### Section C: Exploring Pandas
11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np

# ---- Creating the DataFrame ----
data = {
    "Name": ["John", "Mary", "Alex", "Sara", "Tom",
             "John", "Nina", "Bob", "Alex", "Kim"],
    
    "Age": [25, 30, 22, 28, 35,
            25, 27, 31, 22, 29],
    
    "City": ["Delhi", "Mumbai", "Delhi", "Mumbai", "Chennai",
             "Delhi", "Kolkata", "Chennai", "Delhi", "Mumbai"],
    
    "Salary": [50000, 60000, np.nan, 55000, 70000,
               50000, 62000, np.nan, 54000, 58000]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n")
print(df)

# ---- Using info() ----
print("\n DataFrame Info:\n")
print(df.info())

# ---- Using describe() ----
print("\n DataFrame Description:\n")
print(df.describe())

# ---- Selecting Name and City columns ----
print("\n Selecting Name and City columns :\n")
print(df[["Name", "City"]])

# ---- Dropping the City column ----
aft_drop_city = df.drop("City", axis=1)
print("\n DataFrame after dropping City column:\n")
print(aft_drop_city)

# ---- Filling missing salary values with mean ----
mean_salary = df["Salary"].mean()
print("Mean Salary: ",mean_salary)
df["Salary"].fillna(mean_salary, inplace=True)

print("\n DataFrame after filling missing Salary values:\n")
print(df)

# ---- Removing duplicate rows ----
aft_rem_duplicates = df.drop_duplicates()

print("\n DataFrame after removing duplicates :\n")
print(aft_rem_duplicates)
```
