# Data Visualization
●	Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.
●	Histograms:
	Plot histograms for each numerical column.
	Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.
●	Box Plots:
	Create boxplots for numerical variables to identify outliers and the interquartile range.
	Discuss any findings, such as extreme values or unusual distributions.
●	Bar Chart Analysis for Categorical Column:
	Identify categorical columns in the dataset.
	Create bar charts to visualize the frequency or count of each category.
	Analyze the distribution of categories and provide insights.


```python
# ----------------------------
# 1. Identify column types
# ----------------------------
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

```python
# ----------------------------
# 2. Histograms for Numerical Columns
# ----------------------------
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
index = 1
for col in numerical_cols:
    plt.subplot(3, 3, index)
    plt.hist(df[col], bins=20, edgecolor='black')
    plt.title("Histogram of " + col)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    index += 1

plt.tight_layout()
plt.show()
```

1. Volume
--The histogram shows a right-skewed (positively skewed) distribution.
--A few transactions have very high volume values (around 20–31 units).
--Most values are between 1–6 units, but some high-volume orders appear.
2. Avg Price
--The histogram is highly right-skewed.
--Most products are low to medium priced, but a few very high prices stretch the distribution.
--Several extreme outliers near 40,000–60,000.
--These values are far from the main cluster.
3. Total Sales Value
--Strong positive skew.
--Most sales fall in a lower value range, with a long tail to the right.
--A few transactions have very large sales values, possibly due to:
--High volume
--High price items
4. Discount Rate (%)
--The distribution is slightly left-skewed.
--Most discount rates fall around 14%–20%.
--Very few outliers; values are tightly clustered.
5. Discount Amount
--The histogram is heavily right-skewed.
--Most discounts are small, but some very high discount amounts appear.
--Extreme outliers above 20,000 due to expensive high-value items.
6. Net Sales Value
--Strong positive skew similar to Total Sales Value.
--Most values are low; a few very high values form a long upper tail.
--Several very high-value transactions.

```python
# ------------------------------------
# 2. Boxplots for Numerical Columns
# ------------------------------------
plt.figure(figsize=(15, 10))

index = 1
for col in numerical_cols:
    plt.subplot(3, 3, index)
    plt.boxplot(df[col], vert=False)
    plt.title("Boxplot of " + col)
    plt.xlabel(col)
    index += 1

plt.tight_layout()
plt.show()
```

1. Volume
--The boxplot shows several high-volume outliers (values above 15–30 units).
--Most observations fall in a tight range of 1–6 units, suggesting a clustered distribution.
--These outliers likely represent bulk purchases, special orders, or unusually high demand events.
2. Avg Price
--There are strong positive outliers with prices reaching 40,000–60,000, far above the typical price range.
--The distribution is highly skewed, with most products priced at lower levels.
--This suggests the dataset includes a mix of regular-priced products and a few premium or luxury items.
3. Total Sales Value
--The boxplot reveals many extreme high-value transactions.
--These values occur when:
--High Avg Price combines withn High Volume
4. Discount Rate (%)
--The discount rate distribution is fairly consistent, with minimal outliers.
--Most discounts fall within a narrow range of 14–20%.
--No unusual extreme values appear, showing a stable pricing/discount strategy.
5. Discount Amount
--The boxplot shows several very large discount amounts, some above 20,000.
--These unusual values occur when expensive items receive percentage-based discounts.
--The rest of the data is clustered at much smaller discount amounts, creating a very uneven distribution.
6. Net Sales Value
Similar to Total Sales Value, this variable also shows strong right-skewness with multiple outliers.


```python
# ------------------------------------
# 3. Bar Charts for Categorical Columns
# ------------------------------------
plt.figure(figsize=(15, 12))

index = 1
for col in categorical_cols:
    # Only plot columns with fewer categories (≤20)
    if df[col].nunique() <= 20:
        plt.subplot(4, 3, index)
        df[col].value_counts().plot(kind='bar')
        plt.title("Bar Chart of " + col)
        plt.xlabel(col)
        plt.ylabel("Count")
        index += 1

plt.tight_layout()
plt.show()
```

# Overall Insights from Bar Charts
--Category imbalance is visible—some categories dominate while others have a minimal presence.
--High-frequency categories indicate:
          -Most demanded products
          -Strong-performing markets and brands
--Low-frequency categories may require:
           -Marketing support
           -Better distribution

Bar charts help identify where the business is strong and where improvement is needed.
