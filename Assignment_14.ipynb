### Task 2: Dimensionality Reduction with PCA:
1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.
2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
3.	Transform the original dataset into the principal components.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np

# Standardize features
X = df.drop('Type', axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

# Fit PCA without limiting components
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Scree plot to choose number of components
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

```python

# Select components explaining ~95% variance
n_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.95) + 1
print(f"Number of components explaining 95% variance: {n_components}")

# Transform with selected number of components
pca = PCA(n_components=n_components)
X_pca_reduced = pca.fit_transform(X_scaled)
```

### Task 3: Clustering with Original Data:
1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.
2.	Visualize the clustering results using appropriate plots.
3.	Evaluate the clustering performance using metrics such as silhouette score or Daviesâ€“Bouldin index.


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Apply KMeans on original scaled data
kmeans_orig = KMeans(n_clusters=3, random_state=42)
clusters_orig = kmeans_orig.fit_predict(X_scaled)

# Evaluate clustering
print(f"Silhouette Score (Original): {silhouette_score(X_scaled, clusters_orig):.3f}")
print(f"Davies-Bouldin Index (Original): {davies_bouldin_score(X_scaled, clusters_orig):.3f}")
```

```python
import seaborn as sns
# Visualization of first two PCA components colored by cluster
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters_orig, palette='Set1', legend='full')
plt.title("KMeans Clusters on Original Data (visualized on first two PCA components)")
plt.show()
```

### Task 4: Clustering with PCA Data:
1.	Apply the same clustering algorithm to the PCA-transformed dataset.
2.	Visualize the clustering results obtained from PCA-transformed data.
3.	Compare the clustering results from PCA-transformed data with those from the original dataset.


```python
# Apply KMeans to PCA-reduced data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca_reduced)

# Evaluate clustering
print(f"Silhouette Score (PCA): {silhouette_score(X_pca_reduced, clusters_pca):.3f}")
print(f"Davies-Bouldin Index (PCA): {davies_bouldin_score(X_pca_reduced, clusters_pca):.3f}")

# Visualize clustering on first two principal components
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca_reduced[:, 0], y=X_pca_reduced[:, 1], hue=clusters_pca, palette='Set1', legend='full')
plt.title("KMeans Clusters on PCA-Reduced Data")
plt.show()
```

### Task 5: Comparison and Analysis:
1.	Compare the clustering results obtained from the original dataset and PCA-transformed data.
2.	Discuss any similarities or differences observed in the clustering results.
3.	Reflect on the impact of dimensionality reduction on clustering performance.
4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.


1. Comparison of Clustering Results:
Clustering on the original dataset and on the PCA-transformed dataset usually gives similar overall groupings, especially if PCA retains most of the dataâ€™s variation. The major clusters found by K-means tend to match in both cases, but sometimes PCA can make the group boundaries clearer and reduce noise, helping clusters become more defined.

2. Similarities and Differences:
Both approaches typically find three main clusters in the wine data, which often correspond to the wine types. However, clustering after PCA might shift a few samples into different clusters, as PCA combines related features and removes less important details, sometimes sharpening cluster separation.

3. Impact of Dimensionality Reduction: 
Doing PCA before clustering speeds up the clustering process and can help produce more stable clusters by removing â€œnoiseâ€ and redundancy from the data. However, the principal components are combinations of the original features, so it can be harder to interpret what each cluster means in terms of the raw wine characteristics

4. Trade-offs with PCA vs. Original Data: 
Clustering on original data is easier to explain because clusters are described directly with known features (like alcohol or phenols). Clustering after PCA gives you faster results and can sometimes find better groupings, but the clusters are described by abstract combinations of features, not individual measurements. Use PCA with clustering for large or complex datasets, but use the original data when interpretability is most important.

2.Practical Implications:  
Using PCA before clustering makes data analysis more efficient, especially with large or complex datasets, by reducing computation time and helping clusters emerge more clearly. However, clustering on PCA data can make interpretability harder since the principal components are combinations of original features. Analysts need to balance between efficiency and clarity depending on the analysis goals.

3.Recommendations for Use:  
For datasets with many correlated or redundant features, applying PCA before clustering is recommended to improve performance and reduce noise. When understanding the role of original variables within clusters is important, clustering directly on the original data may be preferable. Overall, PCA is a powerful tool to use when simplifying high-dimensional data for effective clustering.
